#Common methods for selecting data in pandas:

```python
import pandas as pd
#!pip install pandas
```

```python
product_list = {
                'product_name' : ["Red Balloon", "Blue Balloon", "Yellow Balloon", "Black Balloon"],
                'product_id' : ["A100", "B100", "C100", "D100"],
                'product_price' : [1.99, 1.99, 2.49, 3.99]
}
my_products = pd.DataFrame(product_list)
```

```python
my_products
```

##1. Selecting Columns by Name

*   List item
*   List item



```python
# Single Column

my_products['product_price']
```

```python
# Multiple Columns - list

my_products[['product_id', 'product_price']]
```

##2. Selecting All Columns by Index

```python
my_products[0:]
```

# 3. Selecting Columns Using loc

```python
my_products.loc[:, ['product_id', 'product_price']]
```

#4. Selecting Columns Using iloc

```python
my_products.iloc[:,  1:]  # python starts with 0 and the last is not included
```
