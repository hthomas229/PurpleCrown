```python
import pandas as pd
import numpy as np
```

```python
kobe = pd.read_csv("C:\\Users\\hthom\\OneDrive\\R\\kobe.csv")
kobe.head()
```

```python
# Convert the 'date' column to ymd for readability
kobe['date'] = pd.to_datetime(kobe['date'], format='%Y%m%d')
kobe.head()
```

Use Pandas CUM function

```python
# add columns with Kobe's cumulative (running) totals by game
kobe['RunningPts'] = kobe['total_points'].cumsum()  # Cumulative sum
kobe['RunningMax'] = kobe['total_points'].cummax()  # Cumulative maximum
kobe['RunningMin'] = kobe['total_points'].cummin()  # Cumulative minimum
kobe['RunningProd'] = kobe['total_points'].cumprod()  # Cumulative product


kobe
```

```python
kobe
```

Use Expanding

```python
kobe['ExpandingMean'] = kobe['total_points'].expanding().mean()  # Expanding mean
kobe['ExpandingMedian'] = kobe['total_points'].expanding().apply(np.median)
```

```python
kobe
```

1.  Use cum when you want quick cumulative sum max, min product.
2.  Use expanding when you want other custom functions (i.e. mean, median) 
    or to set minimum # of values to use in calc (see min_period arg in rolling  below)

```python
#add a column showing Kobe's total points in a 3-day window
kobe['RollingPts'] = kobe['total_points'].rolling(window=3).sum()
kobe
```

```python

#add column showing Kobe's moving averagee points in a 3 day window
#min_period of 3 says don't start calcualating the mean until you have 3 rows of values
kobe['RollingMean'] = kobe['total_points'].rolling(window=3, min_periods=1).mean()
kobe
```

```python
#center = False is default and sets align right with NAs at the beginning -- current and 2 preceding
#center = True sets align center, NAs at beginiing and end -- current one prceding and one following
kobe['RollingMeanCenter'] = kobe['total_points'].rolling(window=3, center=True).mean()
kobe
```
